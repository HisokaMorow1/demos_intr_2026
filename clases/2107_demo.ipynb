```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from constants import CANONICAL_RESIDUES
from my_functions import get_aa_frequency
```

```python
df_data = pd.read_csv("processed_dataset.csv")
df_data = df_data[["sequence", "label"]]
df_data.head()
```

```python
df_data["label"].value_counts()
```

```python
df_frequencies = get_aa_frequency(df_data)
df_frequencies.head()
```

```python
df_data = df_data.merge(right=df_frequencies, on="sequence")
df_data
```

```python
df_data["length"] = df_data["sequence"].str.len()
```

```python
df_data.describe()
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
scaler = MaxAbsScaler()
scaler.fit(df_data[['Q', 'A', 'W', 'S', 'E', 'D', 'C', 'R', 'F', 'V',
       'T', 'G', 'N', 'H', 'Y', 'M', 'K', 'I', 'L', 'P', 'length']])

df_data_scaled = scaler.transform(df_data[['Q', 'A', 'W', 'S', 'E', 'D', 'C', 'R', 'F', 'V',
       'T', 'G', 'N', 'H', 'Y', 'M', 'K', 'I', 'L', 'P', 'length']])
```

```python
df_data_scaled = pd.DataFrame(data=df_data_scaled, columns=['Q', 'A', 'W', 'S', 'E', 'D', 'C', 'R', 'F', 'V',
       'T', 'G', 'N', 'H', 'Y', 'M', 'K', 'I', 'L', 'P', 'length'])
df_data_scaled.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca_instance = PCA(random_state=42)
pca_instance.fit(df_data_scaled)
```

```python
pca_instance.n_components_, pca_instance.explained_variance_
```

```python
data_transform = pca_instance.transform(df_data_scaled)

df_pca = pd.DataFrame(data=data_transform, columns=[f"PCA_{i+1}" for i in range(pca_instance.n_components_)])
df_pca
```

```python
df_pca["label"] = df_data["label"].values
```

```python
sns.scatterplot(data=df_pca, x="PCA_1", y="PCA_2", hue="label")
```
